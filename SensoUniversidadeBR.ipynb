Em 2016, 34.366 cursos de graduação foram ofertados em 2.407 instituições de educação superior (IES) no Brasil para um total de 8.052.254 estudantes matriculados. Os dados são do Censo da Educação Superior e foram divulgados nesta quinta feira (31) em coletiva de imprensa com a presença do ministro da Educação, Mendonça Filho, da secretária executiva do Ministério da Educação (Mec), Maria Helena Guimarães de Castro, e da presidente do Instituto Nacional de Estudos e Estatísticas Educacionais Anísio Teixeira (Inep), Maria Inês Fini.

Segundo as estatísticas apresentadas, as 197 universidades existentes no país equivalem a 8,2% do total de IES, mas concentram 53,7% das matrículas em cursos de graduação.

No ano passado, o número de matrículas na educação superior (graduação e sequencial) continuou crescendo, mas essa tendência desacelerou quando se comparado aos últimos anos. Entre 2006 e 2016, houve aumento de 62,8%, com uma média anual de 5% de crescimento. Porém, em relação a 2015, a variação positiva foi de apenas 0,2%.

## Cursos

Os cursos de bacharelado mantêm sua predominância na educação superior brasileira com uma participação de 69% das matrículas. Os cursos de licenciatura tiveram o maior crescimento (3,3%) entre os graus acadêmicos em 2016, quando comparado a 2015.

## Vagas

Em 2016, foram oferecidas mais de 10,6 milhões de vagas em cursos de graduação, sendo 73,8% vagas novas e 26,0%, vagas remanescentes. Das novas vagas oferecidas no ano passado, 33,5% foram preenchidas, enquanto apenas 12,0% das vagas remanescentes foram ocupadas no mesmo período.

## Ingressantes

Em 2016, quase 3 milhões de alunos ingressaram em cursos de educação superior de graduação. Desse total, 82,3% em instituições privadas.
Após uma queda observada em 2015, o número de ingressantes teve um crescimento de 2,2% em 2016. Isso ocorreu porque a modalidade a distância aumentou mais de 20% entre os dois anos, enquanto nos cursos presenciais houve um decréscimo no número de ingressantes de 3,7%.

## Concluintes

No ano passado, mais de um 1,1 milhão de estudantes concluíram a educação superior. O número de concluintes em cursos de graduação presencial teve aumento de 2,4% em relação a 2015. A modalidade a distância diminuiu -1,3% no mesmo período.

Entre 2015 e 2016, o número de concluintes na rede pública aumentou 2,9%. Já na rede privada a variação positiva foi de 1,4%. No período de 2006 a 2016, a variação percentual do número de concluintes em cursos de graduação foi maior na rede privada, com 62,6%, enquanto na pública esse crescimento foi de 26,5% no mesmo período.

## Censo da Educação Superior

O Censo da Educação Superior, realizado anualmente pelo Instituto Nacional de Estudos e Pesquisas Educacionais Anísio Teixeira (Inep), constitui-se importante instrumento de obtenção de dados para a geração de informações que subsidiam a formulação, o monitoramento e a avaliação das políticas públicas, além de ser elemento importante para elaboração de estudos e pesquisas sobre o setor. O Censo coleta informações sobre as Instituições de Educação Superior (IES), os cursos de graduação e sequenciais de formação específica e sobre os discentes e docentes vinculados a esses cursos.

Os resultados coletados subsidiam o Sistema Nacional de Avaliação da Educação Superior (Sinaes), seja no cálculo dos indicadores de Conceito Preliminar de Curso (CPC) e do Índice Geral de Cursos (IGC), seja no fornecimento de informações, como número de matrículas, de ingressos, de concluintes, entre outras. As estatísticas possibilitam ainda, através da justaposição de informações de diferentes edições da pesquisa, a análise da trajetória dos estudantes a partir de seu ingresso em determinado curso de graduação, e, consequentemente, a geração de indicadores de acompanhamento e de fluxo na educação superior.

> Preparação e carga de dados em banco para manipulação de informações

```python
#  função que recebe o caminho com arquivo csv, 
#  o nome do banco de dados e o nome da tabela e escreve o conteúdo do csv no banco de dados, na tabela indicada.
#  criação do banco senso com as tabelas aluno, docente, ies, curso, local_oferta, code_brasil e estados
#  alguns .csv muito grandes necessidade de chunk para carregamento da tabela no BD

def create_db_tab(file, name_db, name_tab):

    # Cria/faz conexão com o banco
    db = sqlite3.connect(name_db + '.db')   
    
    # Carrega o arquivo indicado em chunks (lotes) de forma
    # a permitir operação com dataframes grandes demais
    # df_chunks = pd.read_csv(file, chunksize=5000)

    chunksize = 5000
       
    for df_chunk in pd.read_csv(file, low_memory=False, chunksize=chunksize, delimiter="|", encoding='latin-1'):
        df_chunks = []
        df_chunks.append(df_chunk)

        # Para cada lote ("pedaço") do dataframe
        for df in df_chunks:
            # Escreve o pedaço no banco
            df.to_sql(
                name=name_tab, # Define o nome da tabela a receber as informações (ou ser criada)
                index=False, # Não escreve o índice do dataframe na tabela
                if_exists='append', # Se a tabela já existir, adiciona as informações abaixo das existentes
                con=db # Define em qual banco os dados serão inseridos
            )
```

> 1) Crie duas tabelas: uma com número de universidades públicas e privadas por estado e outra com número de alunos por universidade

```python
# 1) Número de universidades públicas e privadas por estado
    # obs.: faculdd especial = 'privada' código '7' da tabela 'ies' 
    # "Especial (art. 242 da Constituição Federal): 
    #  instituição educacional oficial criada por lei estadual ou municipal e existente na data da promulgação 
    #  da Constituição Federal, que não seja total ou preponderantemente mantida com recursos públicos, portanto não gratuita."

query01 = """
        SELECT DISTINCT estados.UF,  
               ( SELECT COUNT(CO_IES) 
                 FROM ies IPub
                 WHERE IPub.TP_CATEGORIA_ADMINISTRATIVA in (1, 2 , 3)
                   AND IPub.CO_UF = ies.CO_UF ) AS Publica,
              ( SELECT COUNT(CO_IES) 
                 FROM ies IPriv
                 WHERE IPriv.TP_CATEGORIA_ADMINISTRATIVA in (4, 5 , 6, 7, 8, 9)
                   AND IPriv.CO_UF = ies.CO_UF ) AS Privada		   
        FROM ies 
        INNER JOIN estados
           ON ies.CO_UF = estados.CO_UF
        ORDER BY estados.UF
            """
df_query01_qtd_priv_pub = pd.read_sql_query(query01, db_senso)
df_query01_qtd_priv_pub
```

```python
# 1) Número de alunos por universidade
query02 = """
        SELECT ies.NO_IES AS Instituicao, count(aluno.ID_ALUNO) AS Qtd_Alunos
        FROM aluno
        INNER JOIN ies
           ON ies.CO_IES = aluno.CO_IES
        GROUP BY Instituicao
        """
df_query01_qtd_alunos_instuicao = pd.read_sql_query(query02, db_senso)
df_query01_qtd_alunos_instuicao
```

> 2) Pergunta-se: é verdade que existem menos mulheres nos cursos de exatas? Explique com os dados.

```python
# 2: é verdade que existem menos mulheres nos cursos de exatas? Explique com os dados.
  # CO_CINE_AREA_GERAL - exatas:
      # 5	Ciências naturais, matemática e estatística
      # 6	Computação e Tecnologias da Informação e Comunicação (TIC)
      # 7	Engenharia, produção e construção

query03 = """

        SELECT 
            CASE 
            WHEN A.TP_SEXO = '1' THEN 'Feminino'
            ELSE 'Masculino'
            END AS Sexo, 
            count(A.ID_ALUNO) AS Qtd_Exatas
        FROM aluno AS A
        inner join code_brasil AS CB
           on A.CO_CINE_ROTULO = CB.CO_CINE_ROTULO 
        WHERE CB.CO_CINE_AREA_GERAL in (5, 6, 7)
        GROUP BY Sexo

          """

df_query_02_qtd_sexo_exatas = pd.read_sql_query(query03, db_senso)
df_query_02_qtd_sexo_exatas
```

```python
# 2 - query alunos NÃO exatas

query04 = """
            SELECT 
                CASE 
                WHEN A.TP_SEXO = '1' THEN 'Feminino'
                ELSE 'Masculino'
                END AS Sexo, 
                count(A.ID_ALUNO) AS Qtd_Nao_Exatas
            FROM aluno AS A
            inner join code_brasil AS CB
               on A.CO_CINE_ROTULO = CB.CO_CINE_ROTULO 
            WHERE CB.CO_CINE_AREA_GERAL not in (5, 6, 7)
            GROUP BY Sexo
         """

df_query_02_qtd_sexo_não_exatas = pd.read_sql_query(query04, db_senso)
df_query_02_qtd_sexo_não_exatas 
```

```python
df_sexo_displinas = pd.concat([df_query_02_qtd_sexo_exatas, df_query_02_qtd_sexo_não_exatas['Qtd_Nao_Exatas']], axis=1)
df_sexo_displinas['Total'] = df_sexo_displinas['Qtd_Exatas'] + df_sexo_displinas['Qtd_Nao_Exatas']
df_sexo_displinas
```

```python
df_perc_disciplinas = df_sexo_displinas[['Qtd_Exatas','Qtd_Nao_Exatas']].apply(lambda x: 100*x/x.sum(), axis=1)
df_perc_disciplinas.index = df_sexo_displinas['Sexo']
df_perc_disciplinas
```

> 4) A afirmação a seguir é verdadeira: Alunos da região Norte têm maior tendência a não concluírem os cursos quando comparados ao restante do país? Comprove com dados.

```python
# Região Norte
query06 = """
            SELECT
                E.UF,
                SUM(QT_MATRICULA_TOTAL) Matriculados,
                SUM(QT_CONCLUINTE_TOTAL) Concluintes,
                SUM(QT_INGRESSO_TOTAL) Ingressantes
            FROM curso C
            INNER JOIN estados E
                    on E.CO_UF = C.CO_UF
            WHERE E.CO_UF in (12, 13, 16, 15, 11, 14, 17)
            GROUP BY E.UF

         """
df_regs_Norte = pd.read_sql_query(query06, db_senso)
df_regs_Norte
```

```python
# proposta de tx desistentes ano Região Norte =
# {[(Ingressantes - Concluintes) / Matriculados] * 100}

taxa_desistAno_Norte = ( (
                        (df_regs_Norte['Ingressantes'].sum() - df_regs_Norte['Concluintes'].sum()) 
                        / df_regs_Norte['Matriculados'].sum() )
                        * 100 )
taxa_desistAno_Norte
```

```python
# Restante do país

# Região Norte
query07 = """
            SELECT
                E.UF,
                SUM(QT_MATRICULA_TOTAL) Matriculados,
                SUM(QT_CONCLUINTE_TOTAL) Concluintes,
                SUM(QT_INGRESSO_TOTAL) Ingressantes
            FROM curso C
            INNER JOIN estados E
                    on E.CO_UF = C.CO_UF
            WHERE E.CO_UF not in (12, 13, 16, 15, 11, 14, 17)
            GROUP BY E.UF

         """
df_regs_RestantePais = pd.read_sql_query(query07, db_senso)
df_regs_RestantePais
```

```python
# proposta de tx desistentes ano regiões diferentes da Norte =
# {[(Ingressantes - Concluintes) / Matriculados] * 100}

taxa_desistAno_RestantePais = ( (
                                (df_regs_RestantePais['Ingressantes'].sum() - df_regs_RestantePais['Concluintes'].sum()) 
                                / df_regs_RestantePais['Matriculados'].sum() )
                                * 100 )
taxa_desistAno_RestantePais
```

> 5) Crie uma variável que represente a taxa de abandono para cada curso. É correto afirmar professores mais/menos capacitados influenciam tal taxa?

```python
# proposta de tx abandono =
# Matriculados T - Concluintes T = Matriculados (T+1) - Ingressantes (T+1)
# Taxa Abandono T = Matriculados T + Concluintes T – Matriculados T+1 + Ingressantes T+1
# TA T = {[(M T – C T – M T+1 + I T+1) / MT] * 100}
```

```python
# inclusão tabela curso de 2019
create_db_tab('SUP_CURSO_2019.csv', 'senso', 'curso_19') 
```

```python
query08 = """
        SELECT
            C18.CO_CURSO AS Codigo,
            C18.NO_CURSO AS NomeCurso,
            C18.QT_MATRICULA_TOTAL AS MatriculadosT,
            C18.QT_CONCLUINTE_TOTAL AS ConcluintesT,
            C19.QT_MATRICULA_TOTAL AS MatriculadosT1,
            C19.QT_INGRESSO_TOTAL AS IngressantesT1
        From curso AS C18
        INNER JOIN curso_19 AS C19
                ON C18.CO_CURSO = C19.CO_CURSO
               AND C18.CO_IES = C19.CO_IES  
         """
df_regs_MatConIng1819 = pd.read_sql_query(query08, db_senso)
df_regs_MatConIng1819.head(10)

# Obs.: curso 2018 38256 linhas / curso 2019 40709 linhas / linhas resultantes do inner join 36770 
```

```python
# incluindo taxa abandono no data frame
df_regs_MatConIng1819['TaxaAbandono'] = ( ( ( df_regs_MatConIng1819['MatriculadosT'] - df_regs_MatConIng1819['ConcluintesT'] 
      - df_regs_MatConIng1819['MatriculadosT1'] + df_regs_MatConIng1819['IngressantesT1'] )
                                         / df_regs_MatConIng1819['MatriculadosT'] ) * 100 )
```

```python
df_regs_MatConIng1819.head(15)
```

```python
# Não achei join docente com o curso e sim com a instituição
# Precisa-se criar taxa abandono pela IES

query09 = """
        SELECT
            ies.NO_IES Instituicao,
            ies.CO_IES,	
            avg(D.TP_ESCOLARIDADE) AS Escolaridade,
            sum(C18.QT_MATRICULA_TOTAL) AS Total_MatriculadosT,
            sum(C18.QT_CONCLUINTE_TOTAL) AS Total_ConcluintesT,
            sum(C19.QT_MATRICULA_TOTAL) AS Total_MatriculadosT1,
            sum(C19.QT_INGRESSO_TOTAL) AS Total_IngressantesT1
        From curso AS C18
        INNER JOIN curso_19 AS C19
                ON C18.CO_CURSO = C19.CO_CURSO 
               AND C18.CO_IES = C19.CO_IES
        INNER JOIN ies 
                ON C18.CO_IES = ies.CO_IES
        INNER JOIN docente D
                ON D.CO_IES = ies.CO_IES
        WHERE D.TP_SITUACAO = 1
        GROUP BY Instituicao, ies.CO_IES
        Order BY Instituicao
            """
df_regs_MatConIng1819_Inst = pd.read_sql_query(query09, db_senso)
df_regs_MatConIng1819_Inst.head(10)

# Obs.: linhas resultantes do inner join 2440 
```

```python
# incluindo taxa abandono no data frame
df_regs_MatConIng1819_Inst['TaxaAbandono'] = ( ( ( df_regs_MatConIng1819_Inst['Total_MatriculadosT'] - df_regs_MatConIng1819_Inst['Total_ConcluintesT'] 
      - df_regs_MatConIng1819_Inst['Total_MatriculadosT1'] + df_regs_MatConIng1819_Inst['Total_IngressantesT1'] )
                                         / df_regs_MatConIng1819_Inst['Total_MatriculadosT'] ) * 100 )
```

> 2) Qual a probabilidade de pegarmos um professor ao acaso no Brasil e ele ter mais do que o dobro da sua idade dado que ele é homem? Dica: procure por Probabilidade Condicional / Teorema de Bayes

```python
# PI > 45 anos
# Quantidade de homens e mulheres professores
query13 =   """
            SELECT 
                CASE 
                WHEN TP_SEXO = 1 THEN 'Feminino'
                ELSE 'Masculino'
                END AS Sexo, 
                count(ID_DOCENTE) AS Qtd_Prof_45M
            FROM docente
            WHERE NU_IDADE > 45 
            GROUP by Sexo
            """
df_Prof_45M = pd.read_sql(query13, db_senso)
df_Prof_45M
```

```python
# P M ou F
query14 = """
            SELECT 
                CASE 
                WHEN TP_SEXO = 1 THEN 'Feminino'
                ELSE 'Masculino'
                END AS Sexo, 
                count(ID_DOCENTE) AS Qtd_Prof_MF
            FROM docente
            GROUP by Sexo
        """
df_Prof_MF = pd.read_sql(query14, db_senso)
df_Prof_MF
```

```python
# Probabilidade de ser sorteado professor sexo masculino após a constatação que é > de 45 anos de idade
# Bayes -> P(M|I>45) = ( ( P(M) . P(I>45|M) ) / P(I>45) )

# probabilidade de ser masculino
PM = df_Prof_MF['Qtd_Prof_MF'][1]/df_Prof_MF['Qtd_Prof_MF'].sum()
# probabilidade de ser acima de 45
PI45 = df_Prof_45M['Qtd_Prof_45M'].sum()/df_Prof_MF['Qtd_Prof_MF'].sum()
# probabilidade de ser masculino com + de 45
PI45M = df_Prof_45M['Qtd_Prof_45M'][1]/df_Prof_MF['Qtd_Prof_MF'][1]

Prob_Masc_IdadeMaior45 = ( ( PM * (PI45M)  ) / PI45 )

Prob_Masc_IdadeMaior45
```

```python
# Calculando a média:
media_show = (1-p)*n
print('média =', media_show)

# Cálculo do desvio-padrão 
std_amostral = (n*p*(1-p))**0.5
print('desvio-padrão =', std_amostral)
```

```python
# intervalo de confiança para 95%
int_inferior = media_show - 1.96*std_amostral
int_superior = media_show + 1.96*std_amostral
print('Intervalo de confiança:', int_inferior, '<-->', int_superior)
```

```python
# cálculo risco financeiro
risco_financeiro = ( num_min_alunos - int_inferior ) * risco_aluno
print ('Risco financeiro total =', risco_financeiro)
```
